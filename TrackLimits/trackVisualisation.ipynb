```python
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
nums = [str(i) for i in range(2,9)]

segments = [np.genfromtxt("ROAD_GR"+n+".csv", dtype=None, delimiter=',') for n in nums]

track = np.concatenate(segments, dtype=float)
```

```python
sns.set_theme()

fig = plt.figure()
ax = plt.axes()
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'pink']

ax.scatter(segments[0].T[0,0:200], segments[0].T[2, 0:200], marker = "o", s = 1)

ax.scatter(segments[0].T[0,10], segments[0].T[2, 10], marker = "o",color='orange', s = 1)

plt.title("Nordschleife")
plt.tight_layout()
plt.xlabel("X (m)")
plt.ylabel("Y (m)")

plt.show()
```
